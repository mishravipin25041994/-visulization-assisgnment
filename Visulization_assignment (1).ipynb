```python
 # 5.Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
```

```python

np.random.seed(0)
data = np.random.randn(100, 5)
df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5'])
corr = df.corr()
corr.loc['Feature1':'Feature5', 'Feature1':'Feature5'] += 0.5
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Synthetic Dataset')
plt.show()

```

```python
 # 2.Using the Student Grades, create a violin plot to display the distribution of scores across different grade categories.
```

```python
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)
```

```python
df
```

```python

fig = go.Figure()

for grade in df['Grade'].unique():
    fig.add_trace(go.Violin(y=df[df['Grade']==grade]['Score'],
                             name=grade,
                             box_visible=True,
                             meanline_visible=True))
fig.update_layout(title='Distribution of Scores Across Grade Categories',
                  yaxis_title='Score',
                  xaxis_title='Grade',
                  violinmode='group')
fig.show()
```

```python
 # 3.Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
```

```python

pivot_table = df.pivot_table(values='Sales', index='Month', columns='Day')
fig = go.Figure(data=go.Heatmap(
    z=pivot_table.values,
    x=pivot_table.columns,
    y=pivot_table.index,
))
fig.update_layout(
    title='Sales Heatmap',
    xaxis_title='Day',
    yaxis_title='Month'
)
fig.show()
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
```

```python

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(
    title='3D Surface Plot',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)
fig.show()
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns
```

```python

np.random.seed(0)
num_points = 50
x = np.random.uniform(0, 10, num_points)
y = np.random.uniform(0, 10, num_points)
sizes = np.random.uniform(5, 15, num_points)
colors = ['#FF0000', '#00FF00', '#0000FF', '#FFFF00', '#00FFFF', '#FF00FF', '#C0C0C0', '#800000', '#808000', '#008000',
          '#800080', '#008080', '#000080', '#FFA07A', '#20B2AA', '#FF6347', '#7B68EE', '#87CEEB', '#32CD32', '#6495ED']
p = figure(title="Random Scatter Plot", x_axis_label='x', y_axis_label='y', width=800, height=400)
for i in range(num_points):
    p.circle(x[i], y[i], size=sizes[i], color=colors[i % len(colors)], alpha=0.6)
show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import LinearColorMapper, ColorBar

# Sample data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Prepare the data for plotting
source = dict(
    x=xx.ravel(),
    y=yy.ravel(),
    value=data_heatmap.ravel()
)

# Create the heatmap
p = figure(width=600, height=400, x_range=(0, 1), y_range=(0, 1),
           title="Heatmap")
p.rect(x='x', y='y', width=1, height=1, source=source,
       fill_color=linear_cmap('value', 'Viridis256', 0, 1),
       line_color=None)

# Add color bar
color_mapper = LinearColorMapper(palette='Viridis256', low=0, high=1)
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
output_notebook()
show(p)
```
